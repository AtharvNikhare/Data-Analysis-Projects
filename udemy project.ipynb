### 7. Now we will list out all the courses of 'Graphic Design' that have price below 100.

```python
df[df['subject'].isin(['Graphic Design'])]
```

##### from above we can see that there are a few rows that have courses free of cost. Due to this engagement of string we won't be able to fetch data according to our requirements.So we will remove the rows containing price as 'Free' and then carry on with our work.

```python
df1=df[df.price!='Free']
df1
```

```python
# here we are changing the datatype of price column to integer as it was in string previously.
df1.price=df1.price.astype(int)
```

```python
#data1[(data1.subject == 'Graphic Design')
df1[(df1.subject=='Graphic Design')&(df1.price<100)].head(60)
```
