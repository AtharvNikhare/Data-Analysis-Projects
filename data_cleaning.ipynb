```python
import pandas as pd
df4=pd.read_csv('E:/data_cleaning/filep.csv')
df4
```

```python
g=df4.groupby('Agency Name')
```

```python
import numpy as np
g['Lifecycle Cost'].agg(np.mean)
```

```python
import matplotlib.pyplot as plt
```

```python
s=df4.groupby(['Agency Name'])['Lifecycle Cost'].size().plot(kind='barh')
plt.xlabel('Cost size')
plt.ylabel('Name')
```

##### the code written below is plotting a graph after sorting out the values obtained after grouping by

```python
s=df4.groupby(['Agency Name'])['Lifecycle Cost'].size().reset_index()
s.sort_values(['Lifecycle Cost'],ascending=False).head(10).plot(kind='barh')
plt.xlabel('Cost Size')
plt.ylabel('Agency Name')
```
